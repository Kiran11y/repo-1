```python
import numpy as np
import pandas as pd
from sklearn.impute import KNNImputer
from sklearn.preprocessing import MinMaxScaler

mm = MinMaxScaler()

def find_category_mappings(df, variable):
    return {k: i for i, k in enumerate(df[variable].dropna().unique(), 0)}

def integer_encode(df , variable, ordinal_mapping):
    df[variable] = df[variable].map(ordinal_mapping)

mappin = dict()
def imputation(df1 , cols):
    df = df1.copy()
    #Encoding dict &amp; Removing nan    
    #mappin = dict()
    for variable in cols:
        mappings = find_category_mappings(df, variable)
        mappin[variable] = mappings

    #Apply mapping
    for variable in cols:
        integer_encode(df, variable, mappin[variable])  

    #Minmaxscaler and KNN imputation 
    sca = mm.fit_transform(df)
    knn_imputer = KNNImputer()
    knn = knn_imputer.fit_transform(sca)
    df.iloc[:,:] = mm.inverse_transform(knn)
    for i in df.columns : 
        df[i] = round(df[i]).astype('int')

    #Inverse transform
    for i in cols:
        inv_map = {v: k for k, v in mappin[i].items()}
        df[i] = df[i].map(inv_map)
    return df


#Imputation
df=imputation(df,['Shape', 'Margin', 'Mass_Density'])
```

```python
y=df['Severity']
X=df.drop('Severity',axis=1)
X.shape,y.shape,df.shape
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.75,random_state=10,stratify=y)
```

```python
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
X_train_rescaled = pd.DataFrame(scaler.fit_transform(X_train), 
                                    columns = X_train.columns, 
                                    index = X_train.index)

X_train_rescaled.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
X_test_rescaled = pd.DataFrame(scaler.fit_transform(X_test), 
                                    columns = X_test.columns, 
                                    index = X_test.index)

X_test_rescaled.head()
```

```python
# model Building
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
knn=KNeighborsClassifier()
knn.fit(X_train_rescaled,y_train)
y_pred=knn.predict(X_test_rescaled)
kn=metrics.classification_report(y_test,y_pred)
print(kn)
```

```python
from sklearn.linear_model import LogisticRegression
regressor=LogisticRegression(C=0.9,penalty='l2')
regressor.fit(X_train_rescaled,y_train)
y_pred=regressor.predict(X_test_rescaled)
lr=metrics.classification_report(y_test,y_pred)
print(lr)
```

```python
from sklearn.svm import SVC
svm=SVC(C=0.3000000,kernel="linear")
svc=svm.fit(X_train_rescaled , y_train)
y_pred=svc.predict(X_test_rescaled)
sv=metrics.classification_report(y_test,y_pred)
print(sv)
```

```python
from sklearn.tree import DecisionTreeClassifier
DTC=DecisionTreeClassifier()
dtc=DTC.fit(X_train_rescaled, y_train)
y_pred=dtc.predict(X_test_rescaled)
dt=metrics.classification_report(y_test,y_pred)
print(dt)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier()
RFC=rfc.fit(X_train_rescaled,y_train)
y_pred=RFC.predict(X_test_rescaled)
rf=metrics.classification_report(y_test,y_pred)
print(rf)
```
